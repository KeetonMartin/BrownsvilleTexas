```python
borderData = pd.read_csv("../input/border-crossing-entry-data/Border_Crossing_Entry_Data.csv")
```

```python
print(borderData.shape)
```

```python
#we have 346733 data entries. Let's narrow that down a bit.
borderData.head()
columns = borderData.columns
print(columns)
brownsvilleData = borderData[borderData['Port Name']=="Brownsville"]
#print(brownsvilleData.head())

brownsvilleTrains = brownsvilleData[brownsvilleData["Measure"] == "Trains"]
print(brownsvilleTrains.head())
print(brownsvilleTrains.shape)
```

```python
import matplotlib.pyplot as plt

brownsvilleTrains.plot(kind='bar', x='Date', 
              y='Value', alpha=0.5, 
              color='darkblue', figsize = (12,9))

plt.title('Trains each month')
plt.xlabel("date")
plt.ylabel("how many trains")

plt.show()
```

```python
recentBrownsvilleTrains = brownsvilleTrains[0:50]

recentBrownsvilleTrains.plot(kind='bar', x='Date', 
              y='Value', alpha=0.5, 
              color='darkblue', figsize = (12,9))

plt.title('Trains each month')
plt.xlabel("date")
plt.ylabel("how many trains")

plt.show()
```
